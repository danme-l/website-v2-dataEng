# Oil Production by Year
[Data from OPEC](https://asb.opec.org/data/ASB_Data.php)

```python
import pandas as pd
import numpy as np
```

```python
oil_df = pd.read_csv('oilProductionByYear.csv')
oil_df.loc[0,'Year'] = 0
oil_df
```

```python
oil_df.loc[:, (oil_df == 'Country').any()]
```

```python
countries = oil_df.loc[:, (oil_df == 'Country').any()]
idx=0
countries.insert(idx, 'Year',value=np.arange(len(countries)))
countries['Year'] = countries['Year'].apply(lambda x: x+1959)
countries = countries.drop(0, axis=0)
```

```python
countries = countries.replace(',','',regex=True)
countries = countries.replace('1','',regex=True)
```

```python
countries = countries.apply(pd.to_numeric)
```

```python
countries
```

```python
countries_long = pd.melt(
    countries,
    id_vars = ['Year'],
    var_name = 'country',
    value_name = 'value'
)
```

```python
# take care of the footnote indicators from the csv file
countries_long = countries_long.replace('1','',regex=True)
```

```python
# add in opec flag
opec_countries = ['Algeria',
'Angola',
'Congo',
'Equatorial Guinea',
'Gabon',
'IR Iran',
'Iraq',
'Kuwait',
'Libya',
'Nigeria',
'Saudi Arabia',
'United Arab Emirates',
'Venezuela']

countries_long['in_opec'] = countries_long['country'].isin(opec_countries).astype(int)
```

```python
lengths = countries_long['country'].str.len()
longest_country = np.where(lengths == lengths.max())[0]
countries_long.iloc[longest_country]
```
